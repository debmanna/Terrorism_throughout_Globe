# The Trend of Terror Attack in India

```python
arr = df1["State"].value_counts().sort_index()

ax = sns.barplot(x = np.array(arr.index), y = np.array(arr.values))
ax.set_xticklabels(ax.get_xticklabels(),fontsize = 10, rotation = 270)
ax.set(xlabel="States -->", ylabel = "Total Number of Terrorists attack from 1970 - 2017",title = "States vs Terror attacks")

plt.show()
```

```python
arr2 = df1["Year"].value_counts().sort_index()

ax2 = sns.barplot(x = np.array(arr2.index), y = np.array(arr2.values))
ax2.set_xticklabels(ax2.get_xticklabels(),fontsize = 10, rotation = 270)
ax2.set(xlabel="Year -->", ylabel = "Total Number of Terrorists attack ",title = "Year vs Terror attacks in India")

plt.show()
```

## Problem 3
Find the frequency of the Casualty in Red Corridor states and in Jammu and Kashmir ?Here Frequency is (Total Casualty/Total Number of a year)

```python
df1_4 = df1[(df1["State"] == "Jharkhand") | (df1["State"] == "Orissa") | (df1["State"] == "Odisha") | (df1["State"] == "Andhra Pradesh") | (df1["State"] == "Chhattisgarh") | (df1["State"] == "Jammu and Kashmir")].copy()

df1_4["Casuality"] = df1_4["Killed"] + df1_4["Wounded"]
```

```python
x = len(df["Year"].unique())
x
```

```python
a = df1_4[df1_4["State"] == "Jammu and Kashmir"]["Casuality"].sum()
s = df1_4["Casuality"].sum()

print("Red Corridor Casuality: ",int((s-a)/x),"Jammu and Kashmir Casuality: ",int(a/x))
```
